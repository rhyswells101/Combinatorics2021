In this file we construct the convex polytopal region denoted $R_f$.

Which are intersections of strips of the form $$b\le Ax \le b+1.$$

We aim to determine the number of cases where $R_f = \emptyset$. 

Along the way we take the opportunity to construct $P_2,P_3,P_4$ and count their respective polytope dimensions.

Consider $P_2$ :

```python
sage: with open(r"C:\Users\RhysL\Sage_Combinatorics_2021\Formated_hyperplanes_P2.txt", "r") as f:
....:     L = f.readlines()
sage: X=[l.strip() for l in L]
empty=[]

for k in range(0,2):
    L_list_line=eval(X[k])
    Poly=Polyhedron(ieqs=L_list_line)
    empty.append(dim(Poly))
print("List of dimensions of regions: ",empty)

print("\n","The number of cases where R_f is dim -1 i.e the empty set is:",empty.count(-1))
print(" The number of cases where R_f is dim 0 is:",empty.count(0))
print(" The number of cases where R_f is dim 1 is:",empty.count(1))
print(" The number of cases where R_f is dim 2 is:",empty.count(2))

print("\n The total number of functions is:",len(empty))
```
